```python
import pandas as pd

df = pd.read_csv('./UFO-completo.csv', error_bad_lines=False)

df
```

```python
#Verifica se tem NaN
hasNan = df.isnull().sum()
print(hasNan)
```

```python
#Remoção das linhas NA
dfCopy = df.copy()

dfCopy.dropna(inplace=True)

dfCopy
```

```python
#Verificação da remoção

hasNan = dfCopy.isnull().sum()
print(hasNan)
```

```python
dfCountry = pd.get_dummies(df['country'], prefix = 'country')

dfCountry
```

```python
# Remoção das colunas que não serão utilizadas
del dfCopy['duration (seconds)']
del dfCopy['duration (hours/min)']
del dfCopy['comments']
del dfCopy['date posted']

dfCopy
```

```python
# Testando uso do LabelEncoder fit
from sklearn.preprocessing import LabelEncoder

city = []
state = []
country = []
shape = []

le = LabelEncoder()
le.fit(dfCopy['city'])
dfCopy['city'] = le.transform(dfCopy['city'])
city = le.inverse_transform(dfCopy['city'])

le.fit(dfCopy['state'])
dfCopy['state'] = le.transform(dfCopy['state'])
state = le.inverse_transform(dfCopy['state'])

le.fit(dfCopy['country'])
dfCopy['country'] = le.transform(dfCopy['country'])
country = le.inverse_transform(dfCopy['country'])

le.fit(dfCopy['shape'])
dfCopy['shape'] = le.transform(dfCopy['shape'])
shape = le.inverse_transform(dfCopy['shape'])

dfCopy
```

```python
print("city: ")
print(city)

print("state: ")
print(state)

print("country: ")
print(country)

print("shape: ")
print(shape)
```

```python
# Não há necessidade de processamento de Outliers

dfCopy[['city', 'state', 'country', 'shape']].plot(kind = 'box')
```
